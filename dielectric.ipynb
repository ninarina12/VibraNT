```python
# Remove outlying tensors
diel_max = 100
crd.data = crd.data[crd.data['diel'].apply(np.abs).apply(np.max) <= diel_max].reset_index(drop=True)
print('Number of samples:', len(crd.data))
```

```python
# Get species
species = sorted(list(set(crd.data['species'].sum())))
n_species = list(np.unique(crd.data['species'].sum(), return_counts=True)[1])
Z_max = max([Atom(k).number for k in species])
print(Z_max)
```

```python
# One-hot encoding atom type and mass
type_encoding = {}
m_specie = []

for Z in tqdm(range(1, Z_max + 1), bar_format=bar_format):
    specie = Atom(Z)
    type_encoding[specie.symbol] = Z - 1
    m_specie.append(atomic_masses[Z])

type_onehot = torch.eye(len(type_encoding))
m_onehot = torch.diag(torch.tensor(m_specie))
```

```python
# Process data into input descriptors
process = Process(species, Z_max, type_encoding, type_onehot, m_onehot, default_dtype)

scale = None   # multiplicative factor for natural cutoff
r_max = 5.     # cutoff radius

tqdm.pandas(desc='Building data', bar_format=bar_format)
crd.data['input'] = crd.data.progress_apply(lambda x: process.build_data(x, r_max, scale), axis=1)
```

```python
# Train/valid/test split
test_size = 0.1
fig = process.train_valid_test_split(crd.data, valid_size=test_size, test_size=test_size, plot=True)
#fig.savefig('images/train_valid_test_split.svg', bbox_inches='tight')
```

```python
# Format dataloaders
batch_size = 16
dataloader_train = tg.loader.DataLoader(crd.data.iloc[process.idx_train]['input'].tolist(), batch_size=batch_size,
                                        shuffle=True)
dataloader_valid = tg.loader.DataLoader(crd.data.iloc[process.idx_valid]['input'].tolist(), batch_size=batch_size)
dataloader_test = tg.loader.DataLoader(crd.data.iloc[process.idx_test]['input'].tolist(), batch_size=batch_size)
```

```python
class E3NN(Network):
    def __init__(self, in_dim, emb_dim, num_layers, max_radius, num_neighbors):
         
        kwargs = {'reduce_output': False,
                  'irreps_in': str(emb_dim)+"x0e",
                  'irreps_out': "1x0e+1x2e",
                  'irreps_node_attr': str(emb_dim)+"x0e",
                  'layers': num_layers,
                  'mul': 32,
                  'lmax': 2,
                  'max_radius': max_radius,
                  'num_neighbors': num_neighbors
                 }
        super().__init__(**kwargs)
        
        # definitions
        self.cmap = cm.lipari
        self.in_dim = in_dim
        self.emb_dim = emb_dim
        self.num_layers = num_layers
        self.max_radius = max_radius
        self.num_neighbors = num_neighbors
        
        self.model_name = 'e3nn_e' + str(emb_dim) + '_l' + str(num_layers)
        
        # embedding
        self.emb_x = nn.Sequential(
            nn.Linear(in_dim, emb_dim),
            nn.ReLU()
        )
        
        self.emb_z = nn.Sequential(
            nn.Linear(in_dim, emb_dim),
            nn.Tanh()
        )
    
    
    def transform(self, data):
        data['x'] = self.emb_x(data['x'])
        data['z'] = self.emb_z(data['z'])
        return super().forward(data)[0]
    
    
    def forward(self, data):
        x = self.transform(data)
        
        # aggregate
        if 'batch' in data:
            batch = data['batch']
        else:
            batch = data['pos'].new_zeros(data['pos'].shape[0], dtype=torch.long)
            
        return torch_scatter.scatter_mean(x, batch, dim=0)
    
    
    def count_parameters(self): 
        return sum(p.numel() for p in self.parameters() if p.requires_grad)
    
    
    def loss(self, y_pred, y_true):
        return nn.MSELoss()(y_pred, y_true)
    
    
    def checkpoint(self, dataloader, device):
        self.eval()
        
        loss_cum = 0.
        with torch.no_grad():
            for j, d in enumerate(dataloader):
                d.to(device)
                y_pred = self.forward(d)
                loss = self.loss(y_pred, d.y).cpu()
                
                loss_cum += loss.item()
                
        return loss_cum/len(dataloader)

    
    def fit(self, opt, dataloader_train, dataloader_valid, history, s0, max_iter=10, device="cpu", scheduler=None):
        chkpt = 1

        for step in range(max_iter):
            self.train()

            loss_cum = 0.
            start_time = time.time()

            for j, d in enumerate(dataloader_train):
                d.to(device)
                y_pred = self.forward(d)
                loss = self.loss(y_pred, d.y).cpu()
                
                print(f"Iteration {step+1:5d}    batch {j+1:5d} / {len(dataloader_train):5d}   " +
                      f"batch loss = {loss.data:.7e}", end="\r", flush=True)
                
                loss_cum += loss.detach().item()
                
                opt.zero_grad()
                loss.backward()
                opt.step()
            
            if scheduler is not None:
                scheduler.step()
            
            end_time = time.time()
            wall = end_time - start_time

            if (step+1)%chkpt == 0:
                print(f"Iteration {step+1:5d}    batch {j+1:5d} / {len(dataloader_train):5d}   " +
                      f"epoch loss = {loss_cum/len(dataloader_train):.7e}")

                loss_valid = self.checkpoint(dataloader_valid, device)
                loss_train = self.checkpoint(dataloader_train, device)

                history.append({
                    'step': step + s0,
                    'wall': wall,
                    'batch': {
                        'loss': loss.item(),
                    },
                    'valid': {
                        'loss': loss_valid,
                    },
                     'train': {
                         'loss': loss_train,
                     },
                })

                yield {
                    'history': history,
                    'state': self.state_dict(),
                    'optimizer': opt.state_dict(),
                    'scheduler': scheduler.state_dict() if scheduler else None
                }
```

```python
args_enn = {'in_dim': Z_max,
            'emb_dim': 64,
            'num_layers': 2,
            'max_radius': r_max,
            'num_neighbors': process.n_train.mean(),
           }

enn = E3NN(**args_enn).to(device)
opt = torch.optim.Adam(enn.parameters(), lr=3e-3)
scheduler = None #torch.optim.lr_scheduler.ExponentialLR(opt, gamma=0.99)
model_path = 'models/' + enn.model_name + '.torch'

print(model_path)
#print(enn)
print('Number of parameters:', enn.count_parameters())
```

```python
fig = enn.visualize()
```

```python
resume = False
    
if resume:
    saved = torch.load(model_path, map_location=device)
    enn.load_state_dict(saved['state'])
    opt.load_state_dict(saved['optimizer'])
    try:
        scheduler.load_state_dict(saved['scheduler'])
    except:
        scheduler = None
    history = saved['history']
    s0 = history[-1]['step'] + 1

else:
    history = []
    s0 = 0
```

```python
# fit E3NN
for results in enn.fit(opt, dataloader_train, dataloader_valid, history, s0, max_iter=20, device=device,
                       scheduler=scheduler):
    with open(model_path, 'wb') as f:
        torch.save(results, f)
```

```python
saved = torch.load(model_path, map_location=device)
history = saved['history']

steps = [d['step'] + 1 for d in history]
loss_train = [d['train']['loss'] for d in history]
loss_valid = [d['valid']['loss'] for d in history]

fig, ax = plt.subplots(figsize=(7,2.5))
ax.plot(steps, loss_train, label='Train', color=process.colors['Train'])
ax.plot(steps, loss_valid, label='Valid.', color=process.colors['Valid.'])
ax.set_xlabel('Iterations')
ax.set_ylabel('Loss')
ax.legend(frameon=False)
#fig.savefig('images/loss_' + enn.model_name + '.svg', bbox_inches='tight')
```

```python
# Predict on all data
enn.eval()
with torch.no_grad():
    tqdm.pandas(desc='Prediction', bar_format=bar_format)
    crd.data['diel_pred'] = crd.data['input'].progress_apply(
        lambda x: process.tij.to_cartesian(enn(x.clone().to(device)).cpu()[0], process.rtp).numpy())
```

```python
# Plot predictions
for k, (idx, dset) in enumerate(zip([process.idx_train, process.idx_valid, process.idx_test],
                                    ['train', 'valid', 'test'])):
    fig, ax = plt.subplots(3,3, figsize=(3*2,3*2), sharex=True, sharey=True)
    fig.subplots_adjust(wspace=0.1, hspace=0.1)

    x = np.stack(crd.data.iloc[idx]['diel'])
    y = np.stack(crd.data.iloc[idx]['diel_pred'])

    d_min = min(x.min(), y.min())
    d_max = max(x.max(), y.max())

    for i in range(3):
        for j in range(3):
            ax[i,j].plot([d_min, d_max], [d_min, d_max], color='black', ls='dashed')
            ax[i,j].scatter(x[:,i,j], y[:,i,j], ec='black', zorder=2, color=list(process.colors.values())[k])
            r2 = max(0., r2_score(x[:,i,j], y[:,i,j]))
            ax[i,j].text(0.1, 0.9, '$R^2$ = {:.2}'.format(r2), ha='left', va='top', transform=ax[i,j].transAxes)
    ax[-1,1].set_xlabel('True')
    ax[1,0].set_ylabel('Predicted')
    fig.savefig('images/prediction_' + enn.model_name + '_' + dset + '.png', dpi=400, bbox_inches='tight')
```
